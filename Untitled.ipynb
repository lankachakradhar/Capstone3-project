```python
import ast

# Function to convert string representations of dictionaries into actual dictionaries
def convert_to_dict(column):
    return column.apply(lambda x: ast.literal_eval(str(x)) if isinstance(x, str) else x)

# Convert relevant columns to dictionaries
df['new_car_detail'] = convert_to_dict(df['new_car_detail'])
df['new_car_overview'] = convert_to_dict(df['new_car_overview'])
df['new_car_feature'] = convert_to_dict(df['new_car_feature'])
df['new_car_specs'] = convert_to_dict(df['new_car_specs'])

# Extracting relevant details from the nested columns
df_extracted = pd.DataFrame()

# Extract details from 'new_car_detail'
df_extracted['fuel_type'] = df['new_car_detail'].apply(lambda x: x.get('ft') if isinstance(x, dict) else None)
df_extracted['body_type'] = df['new_car_detail'].apply(lambda x: x.get('bt') if isinstance(x, dict) else None)
df_extracted['km_driven'] = df['new_car_detail'].apply(lambda x: x.get('km') if isinstance(x, dict) else None)

# Extract details from 'new_car_overview'
df_extracted['registration_year'] = df['new_car_overview'].apply(lambda x: x.get('top')[0]['value'] if isinstance(x, dict) and 'top' in x else None)

# Add the city column
df_extracted['City'] = 'Chennai'

# Display the structured data
df_extracted.head()
```

```python
# Extracting additional fields from 'new_car_detail'
df_extracted['owner_type'] = df['new_car_detail'].apply(lambda x: x.get('ot') if isinstance(x, dict) else None)
df_extracted['insurance_type'] = df['new_car_detail'].apply(lambda x: x.get('it') if isinstance(x, dict) else None)

# Extracting additional fields from 'new_car_overview'
df_extracted['make_model'] = df['new_car_overview'].apply(lambda x: x.get('top')[1]['value'] if isinstance(x, dict) and len(x.get('top', [])) > 1 else None)

# Extract details from 'new_car_feature' (assuming top-level keys like 'value' describe features)
df_extracted['features'] = df['new_car_feature'].apply(lambda x: ', '.join([f['value'] for f in x.get('top', [])]) if isinstance(x, dict) else None)

# Extract details from 'new_car_specs' (e.g., engine type, transmission, etc.)
df_extracted['engine_type'] = df['new_car_specs'].apply(lambda x: x.get('top')[0]['value'] if isinstance(x, dict) and 'top' in x else None)
df_extracted['transmission'] = df['new_car_specs'].apply(lambda x: x.get('top')[1]['value'] if isinstance(x, dict) and len(x.get('top', [])) > 1 else None)

# Retain the car links
df_extracted['car_links'] = df['car_links']

# Display the updated structured data
df_extracted.head()
```
